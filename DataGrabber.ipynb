```python
import imageio

import boto3


ddb=boto3.resource('dynamodb')
table=ddb.Table('vfx_seg')
```

```python
import sys
for data in Samples:

    try:
        img,label=data
        print img
        print label

        dimg=imageio.imread(img)
        print dimg.shape
        
        dlabel=imageio.imread(label)
        print dlabel.shape
        
        item={}
        item['image']=img
        item['label']=label
        item['dim']=list(dimg.shape)
        item['ldim']=list(dlabel.shape)
        
        if dimg.shape==(1080,1920,3) and dlabel.shape==(1080,1920):
            item['train_include']=True
        table.put_item(Item=item)
        
        
    except:
        print  sys.exc_info()[0]
        
        continue
        
        
```
